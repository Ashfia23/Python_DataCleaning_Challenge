```python
import pandas as pd 
```

```python
data_import = pd.read_csv("data_cleaning_challenge.csv")
data_import
```

```python
drop_columns=data_import.drop(columns=["Unnamed: 9","Unnamed: 10"])
drop_columns
```

```python
#renaming column headers
drop_columns.columns=["Row Type","Iter Number","Power1","Speed1","Speed2","Electricity","Effort","Weight","Torque"]
drop_columns.head()
```

```python
#dropping rows like row2 that have header values
drop_columns_header_val= drop_columns[drop_columns["Row Type"]!="Row Type"]
drop_columns_header_val
```

```python
#dropping nan values from Row Type column(NaN are float type comparison with text will give error) 
df=drop_columns_header_val.dropna(subset=["Row Type"])
df
```

```python
#the plan is to add a iterator column for the group data datatest that increments on the next dataset group
column_values=[]
counter=0
for i in df["Row Type"]:
    if "first name" in i:
        counter+=1
    column_values.append(counter)
```

```python
df_iter_col= df
df_iter_col["Iteration"]=column_values
df_iter_col.head(20)
```

```python
name_df=df_iter_col[df_iter_col["Row Type"].str.contains("first name")]
name_df
```

```python
name_df.rename(columns={"Row Type": "First Name","Iter Number": "Last Name","Power1": "Date"}, inplace=True)
name_df
```

```python
df_drop_new=name_df.drop(columns=["Speed1","Speed2","Electricity","Effort","Weight","Torque"])
df_drop_new["First Name"]= df_drop_new["First Name"].str[12:]
df_drop_new["Last Name"]= df_drop_new["Last Name"].str[11:]
df_drop_new["Date"]= df_drop_new["Date"].str[6:]
df_drop_new
```

```python
#We want those line which does not have first name rows i.e exact opposite of name_df for inverse opposite result we use "~"
no_name_df=df_iter_col[~df_iter_col["Row Type"].str.contains("first name")]
no_name_df
```

```python
#Though the first name lasts name and date have same values all over but this is a format to learn
#how to deal with such data in case there is different names and dates
df_merge=pd.merge(left=df_drop_new,right=no_name_df,how="left",on="Iteration")
df_merge
```
